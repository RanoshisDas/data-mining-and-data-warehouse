```python
def armstrong(a):
    t=a
    ln= len(str(a))
    am=0
    while(a!=0):
        dig=int(a%10)
        am+=(dig**ln)
        a=int(a/10)
    if(t==am):
        print("this is a armstrom number.")
    else:
        print('This is not a armstrong number.')

armstrong(153)
```
